```python
# creates an array with a sequence of numbers between -100 and 100 with 0.5 step
x=[]
for i in range(-200, 201):
    x.append(i/2)
```

```python
# creates an array with cosine values for array x
import math
y =[]
for i in x:
    y.append(math.cos(math.radians(i)))
```

```python
# creates an array with sine values for array x
import math
z =[]
for i in x:
    z.append(math.sin(math.radians(i)))
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y, label = "cosine")
plt.plot(x,z, label = "sine")
plt.legend()
plt.xlabel("degrees")
plt.ylabel("sinosoidal curves")
```
